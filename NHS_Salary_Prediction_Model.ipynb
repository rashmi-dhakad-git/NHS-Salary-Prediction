In order to find the correlation between different features and salary, let's first encode the features from categorical to numerical

We will do label encoding for all the features.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
for column in ['job_title','json_employment_type','specialty','employer','department','location']:
    clean_df[column] = label_encoder.fit_transform(clean_df[column])
```

```python
clean_df
```

```python
# Calculate the correlation matrix
correlation_matrix1 = clean_df.corr()
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix1, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

So, from this correlation matrix, there is no relationship between the salary and the other features like: location, job_title, specialty, etc. However, we will still try to find the top feature which may have a relationship with salary feature.

So let's try to see if the other features of the data might have a relationship with the salary features. 

```python
target_variable = 'json_salary_max'

# Get the features with the highest correlation values
top_features = correlation_matrix1[target_variable].abs().sort_values(ascending=False).index[1:3]

# Display the top features
print(f"The top features with the highest correlation with '{target_variable}' are:")
print(top_features)
```

So it seems like the job_title is the feature which affects maximum salary. 

SimilarSimilarly, let's see for minimum salary

```python
target_variable = 'json_salary_min'

# Get the features with the highest correlation values
top_features = correlation_matrix1[target_variable].abs().sort_values(ascending=False).index[1:3]

# Display the top features
print(f"The top features with the highest correlation with '{target_variable}' are:")
print(top_features)
```

Again, it seems that minimum salary is greatly dependent on job_title.

```python
cc=['pay_band','job_title','job_type','json_salary_max','json_salary_min']
result_df=result_df[cc]
```

```python
result_df
```

```python
from sklearn.model_selection import train_test_split

top_features=['pay_band','job_type','job_title']
X = result_df[top_features]
y_min = result_df['json_salary_min']
y_max = result_df['json_salary_max']

# Split data into training and testing sets
X_train, X_test, y_min_train, y_min_test, y_max_train, y_max_test = train_test_split(
    X, y_min, y_max, test_size=0.2, random_state=42
)
```

```python
result_df
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

# Create Decision Tree Regressor models
model_min = DecisionTreeRegressor(random_state=42)
model_max = DecisionTreeRegressor(random_state=42)

# Train the models
model_min.fit(X_train, y_min_train)
model_max.fit(X_train, y_max_train)

# Make predictions
predictions_min = model_min.predict(X_test).ravel()
predictions_max = model_max.predict(X_test).ravel()
```

```python
import matplotlib.pyplot as plt

# Plot for Minimum Salary
plt.figure(figsize=(10, 6))
plt.scatter(y_min_test, predictions_min, color='blue', label='Actual vs Predicted (Min Salary)')
plt.plot([min(y_min_test), max(y_min_test)], [min(y_min_test), max(y_min_test)], linestyle='--', color='red', linewidth=2, label='Perfect Prediction Line')
plt.title('Actual vs Predicted Minimum Salary')
plt.xlabel('Actual Minimum Salary')
plt.ylabel('Predicted Minimum Salary')
plt.legend()
plt.show()

# Plot for Maximum Salary
plt.figure(figsize=(10, 6))
plt.scatter(y_max_test, predictions_max, color='green', label='Actual vs Predicted (Max Salary)')
plt.plot([min(y_max_test), max(y_max_test)], [min(y_max_test), max(y_max_test)], linestyle='--', color='red', linewidth=2, label='Perfect Prediction Line')
plt.title('Actual vs Predicted Maximum Salary')
plt.xlabel('Actual Maximum Salary')
plt.ylabel('Predicted Maximum Salary')
plt.legend()
plt.show()
```

From the graph we can see that for Maximum Salary most of the predicted and actual salaries lie near to the perfect prediction line, where as for minimum salary we see more outliers. One reason could be that the feature showed loose relationship with pay_band which is not a very well indicator for why certain professionals are being paid a certain amount.
