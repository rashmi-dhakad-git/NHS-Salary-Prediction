```python
# aggregate 'location' into broader groups first

# Example: Aggregating 'location' into top N categories
top_n_locations = df_cleaned['location'].value_counts().nlargest(10).index
df_aggregated = df_cleaned.copy()
df_aggregated.loc[~df_aggregated['location'].isin(top_n_locations), 'location'] = 'Other'

# Create a pivot table for the heatmap
heatmap_data = df_aggregated.pivot_table(index='location', columns=None, values=['json_salary_max','json_salary_min'], aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".2f", linewidths=.5)

plt.title('Heatmap: Relationship between Location and Max Salary')
plt.show()
```
