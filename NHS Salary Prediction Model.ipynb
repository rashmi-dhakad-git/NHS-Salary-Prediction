```python
import numpy as np

# Step 1: Initialize Parameters
def initialize_parameters(num_features):
    return np.zeros((num_features, 1)), 0

# Step 2: Define the Hypothesis Function
def hypothesis(X, theta, b):
    return np.dot(X, theta) + b

# Step 3: Define the Cost Function
def cost_function(y_pred, y_actual):
    m = len(y_actual)
    return (1 / (2 * m)) * np.sum((y_pred - y_actual) ** 2)

# Step 4: Gradient Descent
def gradient_descent(X, y, theta, b, learning_rate, num_iterations):
    m = len(y)

    for _ in range(num_iterations):
        # Hypothesis
        y_pred = hypothesis(X, theta, b)

        # Calculate gradients
        d_theta = (1 / m) * np.dot(X.T, (y_pred - y))
        d_b = (1 / m) * np.sum(y_pred - y)

        # Update parameters
        theta -= learning_rate * d_theta
        b -= learning_rate * d_b

    return theta, b

# Step 5: Training the Model
def train_linear_regression(X, y, learning_rate=0.01, num_iterations=1000):
    num_features = X.shape[1]
    theta, b = initialize_parameters(num_features)
    theta, b = gradient_descent(X, y, theta, b, learning_rate, num_iterations)
    return theta, b

# Step 6: Make Predictions
def predict(X, theta, b):
    return hypothesis(X, theta, b)
```
